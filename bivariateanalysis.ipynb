```python
sns.barplot(x='Pclass', y='Age',hue=titanic['Sex'], data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',hue =titanic['Survived'],data=titanic)
```

```python
#distplot
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
#distplot
sns.kdeplot(titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(titanic[titanic['Survived']==1]['Age'])
```

```python
#heatmap(categorical-categorical)
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

```python
#clustermap(cat-cat)
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```
